# List Comprehension 

```python
L1 = [1,2,3,4,5,6,7]
```

```python
L1 = list(range(1,11))
L1
```

```python
#   exp.     itration
L1=[ x        for x in range(1,11)]
L1
```

```python
# sq of a list 1 to 10 
L1 = [x**2 for x in range(1,11)]
L1
```

```python
l = [23,32,23,12,34,25,46,54,2]
L1 = [x for x in l if x%2==0]
L2 = [x for x in l if x%2!=0]
print(L1)
print(L2)
```

```python
# 2. Create a new list that contains only even numbers from a given list.
List= [10, 15, 20, 25, 30]
L1 = [x for x in List if x%2==0]
L1
```

```python
# 3. Extract all names that start with the letter ‘A’ using list comprehension.
names = ["Amit", "Rahul", "Anuj", "Sahil", "Aman"]
result = [name for name in names if name.startswith("A")]
print(result)
```

```python
# 4. Convert all strings in a list to uppercase using list comprehension.
words = ["mayank", "bhatt", "python", "coding"]
result = [w.upper() for w in words]
print(result)
```

```python
# 5. From a list of numbers, create a list of tuples where each tuple contains (number, square).
Lst =  [1, 2, 3, 4, 5]
[(i,i**2) for i in Lst]
```

```python
# tuple
List= [1, 2, 3, 4, 5]
tuple((i,i**2) for i in List)
```

```python
# dictionary
List= [1, 2, 3, 4, 5]
{(i,i**2) for i in List}
```

```python
# set
List= [1, 2, 3, 4, 5]
{(i,i**2) for i in List}
```
