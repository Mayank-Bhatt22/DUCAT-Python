##  1. Introduction to File Handling

File handling is an essential concept in Python programming. It allows us to read from and write to files on the disk. This is useful for storing data permanently, logging information, reading configuration files, and more.

Python provides built-in functions to handle both text and binary files easily.

### Why File Handling?

* Persistent storage of data
* Reading and writing configuration files
* Logging program output
* Handling large data that doesn't fit into memory


##  2. File Modes

File modes determine how a file should be opened ‚Äî for reading, writing, or appending.

| Mode | Description                                           |
| ---- | ----------------------------------------------------- |
| 'r'  | Read (default). File must exist.                      |
| 'w'  | Write. Creates a new file or truncates existing file. |
| 'a'  | Append. Data is added to end of file.                 |
| 'x'  | Create. Fails if file exists.                         |
| 'b'  | Binary mode. e.g. 'rb', 'wb'                          |
| 't'  | Text mode (default). e.g. 'rt', 'wt'                  |
| '+'  | Read and Write mode. e.g. 'r+', 'w+'                  |

###  Example

```python
file1=open("file.txt" ,"w")
file1.write("hello this is my file handeling  using python class. ")
file1.close()
```

```python
file1=open("file.txt" ,"r")
data=file1.read()
file1.close()
print(data)
```
