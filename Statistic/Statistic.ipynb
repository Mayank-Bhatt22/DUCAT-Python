```python
# find 5 no. of summery
minimum = df['total_bill'].min()
maximum = df['total_bill'].max()
Q1 = df['total_bill'].quantile(.25)
Q2 = df['total_bill'].quantile(.50)
Q3 = df['total_bill'].quantile(.75)
IQR = Q3-Q1
L_limit=Q1-1.5*IQR
U_limit= Q3+1.5*IQR
```

```python
print("Minimum: ", minimum)
print("Maximum: ", maximum)
print("Q1: ", Q1)
print("Q2: ", Q2)
print("Q3: ", Q3)
print("L_limit: ", L_limit)
print("U_limit: ", U_limit)
```
