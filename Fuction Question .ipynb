```python
# 6. Filter even numbers using filter
no = [1, 2, 3, 4, 5, 6]
list(filter(lambda x: x % 2 == 0, no))
```

```python
# 9. Filter positive numbers
no = [-1, 0, 2, -3, 4]
Ans = list(filter(lambda x: x > 0, no))
print(Ans)
```

```python
# 10. Filter palindromes
words = ['madam', 'car', 'level']
result = list(filter(lambda x: x == x[::-1], words))
print(result)
```
