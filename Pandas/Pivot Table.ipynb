```python
import pandas as pd

df=pd.read_csv(r"C:\Users\user\OneDrive\Desktop\New Dataset Ducat\data.csv") 
df
```

```python
df["Age_Group"]=df['Age'].apply(lambda x: "Group(20-30)" if x in range(20,31) else
                                "Group(31-40)" if x in range(31,41) else
                                "Group(41-50)" if x in range(41,51) else
                                "Group(51-60)" if x in range(51,61) else
                                "Other"
                               )
```

```python
df.groupby(["City",'Occupation'])['Salary'].sum() 
```

```python
df.pivot_table(index='City',
              values='Salary',
                columns='Occupation',
               aggfunc='sum'             
              )
```

```python
df.pivot_table(index='Age_Group',
              values='Age',
                columns='Occupation',
               aggfunc='count'             
              )
```

```python
df.pivot_table(index='Age_Group',
              values='Salary',
                columns='Occupation',
               aggfunc='max'             
              )
```

```python
df.pivot_table(index='Occupation',
               values='Salary',
               aggfunc=['count','min','max','mean','sum']
              )
```
