```python
import mysql.connector as con
```

```python
connection=con.connect(
    host='localhost',
    user='root',
    password='123456'
)
print('Connection Ok..')
```

```python
cr=connection.cursor()
```

```python
cr.execute('show databases;')
for i in cr.fetchall():
    print(i)
```

```python
cr.execute('use batch4pm;')
for i in cr.fetchall():
    print(i)
```

```python
cr.execute('show tables;')
for i in cr.fetchall():
    print(i)
```

```python
cr.execute('Select * from employee;')
for i in cr.fetchall():
    print(i)
```

```python
import pandas as pd
connection=con.connect(
    host='localhost',
    user='root',
    password='123456',
    database='batch4pm'
)
print('Connection Ok..')
```

```python
pd.read_sql_query("show databases;",connection)
```

```python
pd.read_sql_query("show tables;",connection)
```

```python
pd.read_sql_query("Select * from employee;",connection)
```

```python
pd.read_sql_query("Select * from department;",connection)
```

```python
query='select name,count(*) from employee as e inner join department d on e.DepartmentID=d.DepartmentID group by name;'
pd.read_sql_query(query,connection)
```
