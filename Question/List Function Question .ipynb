```python
# 26. How do you flatten a list of lists? 
lst = [[1, 2], [3, 4], [5, 6]]
flat = [x for sublist in lst for x in sublist]
print(flat)
```

```python
# 30. How do you check if all elements in a list are unique? 
a = [1, 2, 3, 4, 5, 2]
if len(a) == len(set(a)):
    print("All elements are unique")
else:
    print("Not unique")
```
