```python
# Remove all punctuation marks and give a word count dictionary.
text = """
You should not use ChatGPT all the time because it can make you depend on it instead of earning things on your own. ChatGPT gives fast answers,
but sometimes the information may be incomplete or not exactly correct. If you only copy answers, you may not build real understanding, 
problem-solving skills, or confidence. It's better to learn the basics yourself and use ChatGPT only for support — not for everything.
"""
```

```python
# to split 
data=text.split()
print(data)
```

```python
# Which has punctuation marks in the text
punc_data=[]
for i in data:
    if not i.isalpha():
        punc_data.append(i)
print(punc_data)
```

```python
# By chat GPT 
import string
def word_count(text):
    # Remove punctuation
    cleaned_text = text.translate(str.maketrans('', '', string.punctuation))  
    # Convert to lowercase and split into words
    words = cleaned_text.lower().split()
    # Count word frequency
    word_dict = {}
    for w in words:
        word_dict[w] = word_dict.get(w, 0) + 1
    return word_dict
print(word_count(text))
```

```python
# By teacher
def remove_punct(data):
    L=data.split()
    L1=[]
    for word in L:
        if word.isalpha():
            L1.append(word)
        else:
            w=""
            for i in word:
                if i.isalpha():
                    w+=i
                L1.append(w)
    return L1
text1= remove_punct(text)
Dict1={}
for x in text1:
    Dict1[x]=len(x)
print(Dict1)
```

```python
# Lambda Function

# Lambda to remove punctuation from each word
clean_word = lambda w: "".join(filter(lambda ch: ch.isalpha(), w))

# Lambda to split → clean each word → return list
remove_punct = lambda data: list(map(clean_word, data.split()))

# Lambda to create word:length dictionary
word_len_dict = lambda words: {w: len(w) for w in words if w != ""}

# ---- Run everything ----
text1 = remove_punct(text)
Dict1 = word_len_dict(text1)

print(Dict1)
```
