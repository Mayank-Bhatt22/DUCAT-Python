# Lambda Function
Lambda functions, also known as anonymous functions, are small, unnamed functions defined using the lambda keyword. They are often used for short, simple operations that are passed as arguments to higher-order functions.

# Key Points:

1. **Syntax:** lambda arguments: expression

2. **Anonymous:** They do not have a name.

3. **Single Expression:** They are limited to a single expression.

4. **Inline:** They are often used inline, where they are defined at the point of use.

5. **Use Case:** Ideal for simple operations that are used only once or are not worth naming.

# MAP
**Definition:** Applies a function to all the items in an input list and returns an iterator that yields the results.
1. **Syntax:** `map(function, iterable, ...)`
2. **Parameters:**
    - **function:** The function to apply to each element of the iterable.
    - **iterable:** One or more iterable(s) whose elements will be passed to the function.
3. **Returns:** A map object which can be converted to a list, tuple, or another iterable.

```python
lst=["apple","mango","banana","papaya"]
list(map(lambda x:x.upper(),lst))
```

# Filter()

**Definition: Applies a function to each item of an iterable (e.g., list) and returns an iterator yielding those items for which the function returns True**

***Syntax:*** `filter(function, iterable)`

***Parameters:***
- **function:** A function that tests each element of the iterable. It should return True or False.
- **iterable:** The iterable (e.g., list, tuple) to be filtered.
Return Value:

- Returns an iterator that contains only the elements of the iterable for which the function returns True.
Usage:

```python
# find even item
L1=[10,32,55,12,3,16,43]
list(filter(lambda x:x%2==0,L1))
```

```python
# odd item
L1=[10,32,55,12,3,16,43]
list(filter(lambda x:x%2!=0,L1))
```

```python
# More then four words
L1 = ["mayank","amit","sumit","Ram"]
list(filter(lambda x:len(x)>4,L1))
```

```python
list(filter(lambda x:len(x)<4,L1))
```

```python
# filter the name that starts from a 
list(filter(lambda x:x.startswith("a"),L1))
```

# Reduce
**reduce function in Python is part of the functools module and is used to apply a specified function cumulatively to the items of an iterable (like a list) from left to right, so as to reduce the iterable to a single cumulative value.**

***Syntex:*** `from functools import reduce`

***result=*** `reduce(function, iterable, initializer)`

- ***function:*** A function that takes two arguments and returns a single value.

- ***iterable:*** The iterable (like a list or tuple) whose items you want to reduce.

- ***initializer (optional):*** A value to start with; if provided, it will be placed before the items of the iterable in the calculation, and serves as a default when the iterable is empty.

```python
from functools import reduce
```

```python
# sum
L=[32,7,54,212,45,45,32]
reduce(lambda x,y:x+y,L)
```

```python
# max
L=[32,7,54,212,45,45,32]
reduce(lambda x,y:x if x>y else y,L)
```

```python
# min
L=[32,7,54,212,45,45,32]
reduce(lambda x,y:x if x<y else y,L)
```
