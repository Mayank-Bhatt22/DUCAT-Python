# Types of function arguments.

*1. Default Arguments :* When a parameter is given a default value, it's optional to pass that argument when calling the function. 

*2.Positional Arguments :* These are the most common type. Values are passed to the function in the same order as parameters are defined.

*3. Keyword Arguments:* In keyword arguments, we pass values by specifying the parameter name. The order doesn’t matter

*4. Variable-length Arguments* These allow functions to accept any number of arguments.

##### *args (Non-keyword Variable-length)
Used to pass a variable number of positional arguments. It collects them into a tuple.

##### `**kwargs (Keyword Variable-length)
Used to pass a variable number of keyword arguments. It collects them into a dictionary.

```python
# defalut argument
def div(a=0,b=1):
    return a/2
```

```python
def Bill(**item):
    print("***********Bill***********")
    print("item\t\t Amounts")
    total=0
    for i,j in item.items():
        print(i,j, sep="\t\t:")
        total+=j
    print("**************************")
    print('Total Items :', len(item))
    print('Total Amount :', total)
```

```python
Bill(a=1,b=2,c=3,d=4)
```

```python
def bill(**f):
    s=0
    for i in f:
        s+=i
    return s
```

```python
bill(a=1,b=3)
```

```python
# 1. Print type of each argument
def types(*args):
    for a in args:
        print(a, "→", type(a))

types(10, "Hello", [1, 2, 3], 3.14)
```

```python
# 2. Concatenate multiple strings
def concat(*args):
    return " ".join(args)
    
print(concat("Python", "is", "fun"))
```

```python
# 3. Find average of numbers
def avg(*nums):
    return sum(nums) / len(nums) 
    
print(avg(10, 20, 30, 40))
```

```python
# 4. Join characters into a string
def join_chars(*chars):
    return "".join(chars)

print(join_chars('P', 'y', 't', 'h', 'o', 'n'))
```

```python
# 5 Unpack list elements into args
L = [1, 2, 3, 4]
def unpack(*lst):
    for i in lst:
        print(i)
unpack(*L)
```

```python
# 6. Combine two lists using args
def Combine(*lst):
    L=[]
    for i in lst:
        L.extend(i)
    return L
Combine([1,2,3,4],[4,5,6,7])
```
