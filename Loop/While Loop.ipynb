# While Loop

```python
x=0
while x<=10:
    print(x)
    x+=1
```

```python
x = 0
while x<=1:
    print(round(x,2))
    x+=0.1
```

```python
x = 1
while x>=0:
    print(round(x,2))
    x-=0.1
```

```python
s = "mayank"
i = 0
while i<len(s):
    print(s[i])
    i +=1
```

```python
s = "mayank"
i = -1
while i>= -len(s):
    print(s[i])
    i -=1
```

```python
# count of no
n = 234
counter=0
while n>0:
    counter +=1
    n//=10
print(counter)
```

```python
# sum of no 
n = 435
Sum = 0
while n>0:
    Sum += n%10
    n//=10
print(Sum)
```

```python
n = 123456
rev = 0
while n > 0:
    digit = n % 10        
    rev = rev * 10 + digit  
    n = n // 10           
print("Reversed number:", rev)
```
