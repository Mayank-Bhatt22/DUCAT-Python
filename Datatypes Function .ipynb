

###  Python String Functions with Examples:

| Function | Description | Example |
|----------|-------------|---------|
| `capitalize()` | Capitalizes the first character | `"hello".capitalize()` → `'Hello'` |
| `casefold()` | Case-insensitive lowercase (stronger than `lower()`) | `"HELLO".casefold()` → `'hello'` |
| `center(width)` | Centers the string with padding | `"hi".center(6, '*')` → `'**hi**'` |
| `count(sub)` | Counts how many times a substring occurs | `"banana".count('a')` → `3` |
| `encode()` | Returns encoded version of string | `"café".encode()` → `b'caf\xc3\xa9'` |
| `endswith(suffix)` | Checks if string ends with given value | `"hello".endswith('lo')` → `True` |
| `expandtabs(n)` | Replaces `\t` with spaces | `"a\tb".expandtabs(4)` → `'a   b'` |
| `find(sub)` | Finds the first occurrence index | `"hello".find('l')` → `2` |
| `format()` | Formats string with placeholders | `"{} is {}".format("Sky", "blue")` → `'Sky is blue'` |
| `format_map(dict)` | Formats using a mapping | `"{x}, {y}".format_map({'x': 1, 'y': 2})` → `'1, 2'` |
| `index(sub)` | Like `find()`, but raises error if not found | `"hi".index('i')` → `1` |
| `isalnum()` | Checks if alphanumeric | `"abc123".isalnum()` → `True` |
| `isalpha()` | Checks if only alphabets | `"abc".isalpha()` → `True` |
| `isascii()` | Checks if all characters are ASCII | `"hello".isascii()` → `True` |
| `isdigit()` | Checks if all characters are digits | `"123".isdigit()` → `True` |
| `islower()` | Checks if all lowercase | `"hello".islower()` → `True` |
| `isupper()` | Checks if all uppercase | `"HELLO".isupper()` → `True` |
| `istitle()` | Checks if title case | `"Hello World".istitle()` → `True` |
| `isspace()` | Checks if only whitespace | `" \t".isspace()` → `True` |
| `join(iterable)` | Joins iterable elements into string | `"-".join(["a", "b"])` → `'a-b'` |
| `ljust(width)` | Left-justifies string | `"hi".ljust(5, '_')` → `'hi___'` |
| `lower()` | Converts to lowercase | `"Hi".lower()` → `'hi'` |
| `lstrip()` | Removes leading spaces/chars | `"  hello".lstrip()` → `'hello'` |
| `partition(sep)` | Splits into 3 parts: before, sep, after | `"apple:fruit".partition(':')` → `('apple', ':', 'fruit')` |
| `replace(old, new)` | Replaces substrings | `"hello".replace('l', 'x')` → `'hexxo'` |
| `rfind(sub)` | Finds last occurrence index | `"hello".rfind('l')` → `3` |
| `rindex(sub)` | Like `rfind()`, but raises error if not found | `"hello".rindex('l')` → `3` |
| `rjust(width)` | Right-justifies string | `"hi".rjust(5, '_')` → `'___hi'` |
| `rstrip()` | Removes trailing spaces/chars | `"hello  ".rstrip()` → `'hello'` |
| `rsplit(sep)` | Splits from the right | `"a,b,c".rsplit(',', 1)` → `['a,b', 'c']` |
| `split(sep)` | Splits into list | `"a,b,c".split(',')` → `['a', 'b', 'c']` |
| `splitlines()` | Splits at line breaks | `"a\nb".splitlines()` → `['a', 'b']` |
| `startswith(prefix)` | Checks if starts with value | `"hello".startswith('he')` → `True` |
| `strip()` | Removes leading/trailing spaces/chars | `"  hi  ".strip()` → `'hi'` |
| `swapcase()` | Swaps upper/lowercase | `"HeLLo".swapcase()` → `'hEllO'` |
| `title()` | Converts to title case | `"hello world".title()` → `'Hello World'` |
| `translate(map)` | Translates using translation table | `"abc".translate(str.maketrans('a', 'x'))` → `'xbc'` |
| `upper()` | Converts to uppercase | `"hi".upper()` → `'HI'` |
| `zfill(width)` | Pads string on left with zeros | `"42".zfill(5)` → `'00042'` |

## Built-in Functions Used with Lists

These are general-purpose Python functions you can use on lists:

| Function      | Description                   | Example                                             |
| ------------- | ----------------------------- | --------------------------------------------------- |
| `len()`       | Returns number of elements    | `len([1, 2, 3]) → 3`                                |
| `max()`       | Returns the largest item      | `max([1, 5, 3]) → 5`                                |
| `min()`       | Returns the smallest item     | `min([1, 5, 3]) → 1`                                |
| `sum()`       | Returns sum of elements       | `sum([1, 2, 3]) → 6`                                |
| `sorted()`    | Returns a sorted list         | `sorted([3, 1, 2]) → [1, 2, 3]`                     |
| `list()`      | Converts iterable to list     | `list("abc") → ['a', 'b', 'c']`                     |
| `any()`       | True if any element is true   | `any([0, 1, False]) → True`                         |
| `all()`       | True if all elements are true | `all([1, True, 5]) → True`                          |
| `enumerate()` | Returns index-value pairs     | `list(enumerate(['a','b'])) → [(0, 'a'), (1, 'b')]` |
| `zip()`       | Combines multiple lists       | `list(zip([1,2],[3,4])) → [(1,3), (2,4)]`           |

## List Methods

These are methods that only work on list objects:

| Method             | Description              | Example                                 |
| ------------------ | ------------------------ | --------------------------------------- |
| `append(x)`        | Adds item at the end     | `a = [1]; a.append(2) → [1, 2]`         |
| `extend(iterable)` | Adds multiple items      | `a = [1]; a.extend([2,3]) → [1, 2, 3]`  |
| `insert(i, x)`     | Inserts at index `i`     | `a = [1,3]; a.insert(1, 2) → [1, 2, 3]` |
| `remove(x)`        | Removes first occurrence | `a = [1,2,3]; a.remove(2) → [1, 3]`     |
| `pop([i])`         | Removes and returns item | `a = [1,2,3]; a.pop() → 3`              |
| `clear()`          | Removes all items        | `a = [1,2]; a.clear() → []`             |
| `index(x)`         | Returns index of x       | `[10, 20, 30].index(20) → 1`            |
| `count(x)`         | Counts occurrences       | `[1,2,2,3].count(2) → 2`                |
| `sort()`           | Sorts the list in place  | `a = [3,1,2]; a.sort() → [1, 2, 3]`     |
| `reverse()`        | Reverses the list        | `a = [1,2,3]; a.reverse() → [3, 2, 1]`  |
| `copy()`           | Returns shallow copy     | `b = a.copy()`                          |

---

###  Notes:

* `sort()` modifies the original list; `sorted()` returns a new sorted list.
* `pop()` without arguments removes the **last** item.
* `remove(x)` gives an error if `x` not found.
* `append()` adds a **single item**, even if it’s a list.

##  **2. Set Built-in Functions**

| `any(set)`       | Returns `True` if any element is true   | `any({0, 0, 3})  # True`         |
| `enumerate(set)` | Returns enumerate object                | `list(enumerate({10, 20, 30}))`  |
