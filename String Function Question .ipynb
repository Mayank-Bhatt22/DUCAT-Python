```python
# 1. Write a Python function to reverse a string. 
def reverse_string(s):
    return s[::-1]
print(reverse_string("mayank"))
```

```python
# 2. Write a Python function to count the number of vowels in a string. 
def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for char in s if char in vowels)
print(count_vowels("mayank bhatt"))
```

```python
# 3. Write a Python function to check whether a string is a palindrome. 
is_palindrome = lambda s: s.lower().replace(" ", "") == s.lower().replace(" ", "")[::-1]
print(is_palindrome("madam"))
print(is_palindrome("Mayank"))
```

```python
# 7. Write a Python function to find the length of the longest word in a string. 
a = ("mayank","bhatt","mayankbhatt")
longest_word = lambda words: max(words, key=len)
print(longest_word(a))
```

```python
# 8. Write a Python function to replace a specified phrase with another specified phrase in a string. 
def replace_phrase(text, old, new):
    return text.replace(old, new)
print(replace_phrase("I love Python", "Python", "Java"))
```

```python
# 9. Write a Python function to check if a string starts with a specified prefix. 
starts_with = lambda text, prefix: text.startswith(prefix)
print(starts_with("dexant coding", "dex"))
print(starts_with("dexant coding", "code"))
```

```python
# 10. Write a Python function to check if a string ends with a specified suffix. 
starts_with = lambda text, prefix: text.endswith(prefix)
print(starts_with("dexant coding", "dex"))
print(starts_with("dexant coding", "coding"))
```

```python
# 11. Write a Python function to split a string at each space and return a list of the words. 
split_words = lambda text: text.split()
print(split_words("hello world this is lambda"))
```

```python
# 12. Write a Python function to join a list of words into a single string with spaces in between. 
def join_words(words):
    return " ".join(words)
print(join_words(["mayank", "bhatt", "is", "coding"]))
```

```python
# 13. Write a Python function to remove all whitespace from a string. 
a = "  mayank   bhatt  "
remove_whitespace = lambda s: "".join(s.split())
print(remove_whitespace(a))
```

```python
# 14. Write a Python function to remove a specified character from a string.
def remove_char(s, ch):
    return s.replace(ch, "")
print(remove_char("mayank bhatt", "a"))
```

```python
# 15. Write a Python function to replace all occurrences of a specified character in a string with another character. 
def replace_all(s, old, new):
    return s.replace(old, new)
print(replace_all("mayank bhatt", "a", "x"))
```

```python
# 16. Write a Python function to check if a string contains only alphabetic characters.
def bhai(a):
    return a.isalpha()
print(bhai("1"))
print(bhai("Mayank"))
```

```python
# 17. Write a Python function to check if a string contains only numeric characters. 
def bhai(a):
    return a.isdigit()
print(bhai("1"))
print(bhai("Mayank"))
```

```python
# 18. Write a Python function to count the number of occurrences of a specified character in a string. 
def count_char(s, ch):
    return s.count(ch)
print(count_char("mayank bhatt", "a"))
```

```python
# 19. Write a Python function to find the first occurrence of a specified character in a string. 
def first_occurrence(s, ch):
    return s.find(ch)
print(first_occurrence("mayank bhatt", "a"))
```

```python
# 20. Write a Python function to find the last occurrence of a specified character in a string. 
def last_occurrence(s, ch):
    return s.rfind(ch)
print(last_occurrence("mayank bhatt", "a"))
```
