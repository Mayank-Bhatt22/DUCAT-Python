##  Decorator in Python?

A **decorator** is a **function that modifies another function or method without changing its source code**.

 **In simple words:**

> A decorator **wraps** a function to add extra behavior **before or after** it runs.


### Why decorators are needed?

* Logging
* Authentication & authorization
* Performance measurement
* Caching
* Validation
* Access control
* Code reusability (DRY principle)


##  Basic Decorator (Step-by-Step)

### Step 1: Normal function

```python
def say_hello():
    print("Hello")
```

---

### Step 2: Create decorator

```python
def my_decorator(func):
    def wrapper():
        print("Before function call")
        func()
        print("After function call")
    return wrapper
```

---

### Step 3: Apply decorator

```python
@my_decorator
def say_hello():
    print("Hello")
```

````python
# 1st step Normal Function
def say_hello():
    print('Hello')
````

````python
# 2nd step Create decorator
def my_decorator(func):
    def wrapper():
        print('Before funtions call')
        func()
        print('After funtions call')
    return wrapper
````

````python
# 3rd step Apply decorator
@my_decorator
def say_hello():
    print('Hello')
````

````python
say_hello()
````

````python
import time

def time_calculator(func):
    def wrapper(n):
        start_time = time.time()
        print("Starting time:", start_time)

        func(n)

        end_time = time.time()
        print("Ending time:", end_time)
        print("Execution time:", end_time - start_time)
    return wrapper

@time_calculator
def ending_time(n):
    s = 0
    for i in range(n):
        s += i
    print("Sum:", s)
````

````python
ending_time(1000000)
````

````python
import time
def timer(fn):
    def wrapper(*a,**kw):
        start=time.time()
        result=fn(*a,**kw)
        end=time.time()
        print('Execution time:',end-start)
        return result

    return wrapper
@timer
def sum_no(n):
    s=0
    for i in range(n):
        s+=i
    return s
````

````python
sum_no(1000000)
````

````python
@timer
def Sleeps():
    time.sleep(2)
    return ('hello')
````
