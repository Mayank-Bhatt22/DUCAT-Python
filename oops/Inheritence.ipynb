```python
class Animal:
    def __init__(self ,name,age ,color):
        self.name=name
        self.age=age
        self.color=color

    def eat(self):
        print(f"{self.name} is eating...")

    def sleep(self):
        print(f'{self.name} is sleeping...')
    
    def detail(self):
        print("Name  :",self.name)
        print("Age   :",self.age)
        print("Color :",self.color)
```

```python
class Dog(Animal):
    def __init__(self ,name,age ,color):
        super().__init__(name,age ,color)
```

```python
D1=Dog("abc",2,'black')
```

```python
D1.sleep()
```

```python
D1.eat()
```

```python
D1.name
```

```python
D1.detail()
```

```python
class Cat(Animal):
    def __init__(self ,name,age ,color,sound):
        super().__init__(name,age ,color)
        self.sound=sound
    def voice(self):
        print(f"Sound : {self.sound}")
```

```python
C1=Cat('xyz',3, 'white','meow')
```

```python
C1.detail()
```

```python
C1.voice()
```

```python
class Father:  # class
    def __init__(self,f_name,surname,occupation):    # cunstruction
        self.f_name=f_name   # attribute
        self.surname=surname   # attribute
        self.occupation=occupation   # attribute

# self: self us refrance of current variable/instance
    
    def F_detail(self):   # method 
        print("Name  :",self.f_name)
        print("Surname   :",self.surname)
        print("occupation :",self.occupation)
```

```python
class Mother(Father):
    def __init__(self,m_name,m_age,f_name,surname,occupation):
        super().__init__(f_name,surname,occupation)  # refrence classs of parent class
        self.m_name=m_name
        self.m_age=m_age 
    def M_detail(self):
        print("Name  :",self.m_name)
        print("Surname   :",self.surname)
        print("Husband :",self.f_name)  
```

```python
m1=Mother("salini",30,'shankar',"verma","job")
```

```python
m1.M_detail()
```

```python
m1.F_detail()
```

```python
class Father:  # class
    def __init__(self,f_name,surname,occupation):    # cunstruction
        self.f_name=f_name   # attribute
        self.surname=surname   # attribute
        self.occupation=occupation   # attribute

# self: self us refrance of current variable/instance
    
    def F_detail(self):   # method 
        print("Name  :",self.f_name)
        print("Surname   :",self.surname)
        print("occupation :",self.occupation)
```

```python
class Mother:
    def __init__(self,m_name,m_age,m_occupation):
        self.m_name=m_name
        self.m_age=m_age 
        self.m_occupation=m_occupation
        
    def M_detail(self):
        print("Name  :",self.m_name)
        print('Age  :',self.m_age)
        print('Occupation  :',self.m_occupation)
```

```python
m1=Mother('Mayank_ke_mummy',20,'house')
```

```python
m1.M_detail()
```

```python
class Son(Father,Mother):
    def __init__(self,name,age,f_name,surname,occupation,m_name,m_age,m_occupation):
        Father.__init__(self,f_name,surname,occupation)
        Mother.__init__(self,m_name,m_age,m_occupation)
        self.name=name
        self.age=age

    def S_detail(self):
        print('Name :',self.name)
        print('Surname :',self.surname)
        print('Father Name :',self.f_name)
        print('Mother Name :',self.m_name)
```

```python
s1=Son('mayank',15,'kishan','singh','Job','Santi',35,'House wife')
```

```python
s1.F_detail()
```

```python
s1.M_detail()
```

```python
s1.S_detail()
```

# Encapsulation
1. Public (anywhere)
2. Protected (class + subclasses)
3. Private (only within the class) access modifiers

```python
class Person:
    def __init__(self,name,age,salary):
        self.name=name # Public
        self._age=age  # Protected
        self.__salary=salary # Private
    def get_age(self):
        print(self._age)
    def get_salary(self):
        print(self.__salary)
```

```python
p1=Person('Mayank',20,25000)
```

```python
p1.name
```

```python
p1._age
```

```python
p1.get_salary()
```
